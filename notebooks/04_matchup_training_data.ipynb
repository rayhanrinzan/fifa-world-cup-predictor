```python
import pandas as pd
import numpy as np
```

```python
team_matches = pd.read_csv("../data/processed/team_matches_with_form.csv")
team_matches.head()
```

```python
team_matches["date"] = pd.to_datetime(team_matches["date"])
team_matches.columns
```

```python
home_rows = team_matches[team_matches["is_home"] == True].copy()
away_rows = team_matches[team_matches["is_home"] == False].copy()
home_rows.shape, away_rows.shape
```

```python
form_features = [
    "last_5_points_per_match",
    "last_5_goals_for_per_match",
    "last_5_goals_against_per_match",
    "last_5_goal_difference_per_match",
    "last_5_win",
    "last_5_draw",
    "last_5_loss"
]
```

```python
[col for col in team_matches.columns if "last_5" in col]
```

```python
team_a = home_rows[
    [
        "date",
        "team",
        "opponent",
        "goals_for",
        "goals_against",
        "result",
        "tournament",
        "neutral",
    ] + form_features
].copy()
```

```python
team_a = team_a.rename(columns={
    "team": "team_a",
    "opponent": "team_b",
    "goals_for": "team_a_goals",
    "goals_against": "team_b_goals",
    "result": "team_a_result"
})
```

```python
team_a = team_a.rename(columns={
    col: f"team_a_{col}" for col in form_features
})
```

```python
team_a.head()
```

```python
team_b = away_rows[
    [
        "date",
        "team",
        "opponent",
    ] + form_features
].copy()
```

```python
team_b = team_b.rename(columns={
    "team": "team_b",
    "opponent": "team_a"
})
```

```python
team_b = team_b.rename(columns={
    col: f"team_b_{col}" for col in form_features
})
```

```python
team_b.head()
```

```python
matchups = team_a.merge(
    team_b,
    on=["date", "team_a", "team_b"],
    how="inner"
)
```

```python
matchups.head()
```

```python
matchups.shape
```

```python
for col in form_features:
    matchups[f"{col}_diff"] = (
        matchups[f"team_a_{col}"] - matchups[f"team_b_{col}"]
    )
```

```python
[col for col in matchups.columns if "diff" in col]
```

```python
matchups["target"] = matchups["team_a_result"]
matchups["target"].value_counts()
```

```python
metadata_columns = [
    "date",
    "team_a",
    "team_b",
    "team_a_goals",
    "team_b_goals",
    "tournament",
    "neutral",
    "target"
]

team_a_feature_columns = [f"team_a_{col}" for col in form_features]
team_b_feature_columns = [f"team_b_{col}" for col in form_features]
diff_feature_columns = [f"{col}_diff" for col in form_features]

final_columns = (
    metadata_columns
    + team_a_feature_columns
    + team_b_feature_columns
    + diff_feature_columns
)

matchup_training_data = matchups[final_columns].copy()
```

```python
matchup_training_data.head()
```

```python
matchup_training_data.to_csv(
    "../data/processed/matchup_training_data.csv",
    index=False
)
```
